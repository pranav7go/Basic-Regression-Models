**Train test split**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.08)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train= sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

```
np.set_printoptions(suppress=True) 
print(X_train[:10])
```

```
print(y_train[:10])
```

## Training the SVR model on the whole dataset

```
from sklearn.ensemble import AdaBoostRegressor
regressor = AdaBoostRegressor()
regressor.fit(X_train, y_train)
```

## Predicting a new result

```
ypred=sc_y.inverse_transform(regressor.predict((X_train)))
```

**R2 score**

```
from sklearn.metrics import r2_score
r2_score((y_test),sc_y.inverse_transform(regressor.predict(sc_X.transform(X_test))))
```

```
r2_score(sc_y.inverse_transform(y_train),ypred)
```
